# 5)

```python
car_sales['Price']=car_sales['Price'].str.replace("[\$\,]", "")
car_sales.head()
```

```python
car_sales.info()
```

```python
car_sales['Price']=car_sales['Price'].astype('int64')
```

```python
car_sales.info()
```

# 6)

```python
car_sales['Total Sales']=car_sales['Price'].cumsum()
car_sales.head()
```

# 7)

```python
car_sales['Sale Date']= pd.date_range(start='2023-02-15', 
                                             periods=len(car_sales))
car_sales.head()
```

# 8)

```python
line_plot=plt.plot(car_sales['Price'],
                   car_sales['Total Sales'], 
                   )
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.title('Price vs Total Sales Line Plot')

plt.show()
```

# 9)

```python
Scatter=plt.scatter(car_sales['Odometer (KM)'], 
                    car_sales['Price'])
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.title('Odometer vs Price Scatter Plot')

plt.show()
```

# 10)

```python
plt.bar(car_sales['Make'], 
        car_sales['Odometer (KM)'])
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.title('Make vs Odometer Bar Plot')

plt.show()
```

# 11)

```python
plt.hist(car_sales['Make'])
plt.xlabel('Make')
plt.title('Histogram of Car Make')

plt.show()
```

# 12)

```python
fig=plt.figure(figsize=(12,10))

ax1=fig.add_subplot(221) #two row, two collumn, first axis 
ax1.set_title('Price vs Total Sales Line Plot')

ax1.plot(car_sales['Price'],
         car_sales['Total Sales'],
        )


ax2=fig.add_subplot(222) #two row, two collumn, first axis 
ax2.set_title('Odometer vs Price Scatter Plot')

ax2.scatter(car_sales['Odometer (KM)'],
         car_sales['Price'],
        )


ax3=fig.add_subplot(223) #two row, two collumn, first axis 
ax3.set_title('Make vs Odometer Bar Plot')

ax3.bar(car_sales['Make'],
         car_sales['Odometer (KM)'],
        )


ax4=fig.add_subplot(224) #two row, two collumn, first axis 
ax4.set_title('Histogram of Car Make')

ax4.hist(car_sales['Make'],
        )

plt.show()
```

# 16)

```python
corr=car_sales[['Total Sales','Price', 'Odometer (KM)']].corr(method='pearson')

cols=['Total Sales','Price', 'Odometer (KM)']
heat_map=sns.heatmap(corr, 
                     annot=True, 
                     xticklabels=cols, 
                     yticklabels=cols,
                     cmap='Blues')
```
